```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
month = ["Jan","Feb","Mar","Apr"]
sales = [100,150,80,130]


plt.subplot(2,2,1)
plt.plot(month,sales,"r",linestyle="--",marker = "*",markerfacecolor='yellow',markersize=10)
for x,y in zip(month,sales):
    plt.text(x,y,y,color="k",ha="center",va="bottom")

plt.subplot(2,2,4)
plt.scatter(month,sales, color = "blue")

plt.subplot(2,2,3)
plt.bar(month,sales, width = 0.3, color = "purple")
for x,y in zip(month,sales):
    plt.text(x,y,y,color="g",ha="center",va="bottom")

plt.subplot(222)
plt.pie(sales, labels = month, autopct='%1.1f%%')


plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("iris_add.csv")
df.head()
df.info()
```

```python
df["date"] = pd.date_range("2024-07-27",periods=150)
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python


plt.figure(figsize=(10,10))
plt.plot(df["date"],df["sepal_aspect"], label = "sepal_aspect_ratio")
plt.plot(df["date"],df["pedal_aspect"], label = "pedal_aspect_ratio")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.hist(df["sepal_aspect"],bins=30,color="r", alpha=0.7, label = "sepal_aspect", width=0.05)
plt.legend()

plt.subplot(2,2,2)
plt.hist(df["pedal_aspect"],bins=30,color="b", alpha=0.7, label = "pedal_aspect",width=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
